### Log all datasets

```python
dataset_a=project.log_dataset(key='model_a_data',df=model_a_input,format='csv')
dataset_b=project.log_dataset(key='model_b_data',df=model_b_input,format='csv')
dataset_c=project.log_dataset(key='model_c_data',df=model_c_input,format='csv')
```

```python
auto_trainer = mlrun.code_to_function("auto_trainer",kind='job',image='mlrun/mlrun',filename='./code/auto_trainer.py')
```

## Creating models with Status=Challenger

```python
model_name = 'Model-A'
```

```python
train_run_a = auto_trainer.run(
        inputs={"dataset": dataset_a.uri},
        params = {
            "model_class": 'sklearn.ensemble.RandomForestClassifier',
            "label_columns": "label",
            "model_name": model_name,
            "labels" : {'Status':'Challenger'},
            "tag" : "v1"
        }, 
        handler='train',
        local=False
    )
```

```python
model_name = 'Model-B'
```

```python
train_run_b = auto_trainer.run(
        inputs={"dataset": dataset_b.uri},
        params = {
            "model_class": 'sklearn.ensemble.RandomForestClassifier',
            "label_columns": "label",
            "model_name": model_name,
            "labels" : {'Status':'Challenger'},
            "tag" : "v1"
        }, 
        handler='train',
        local=False
    )
```

```python
model_name = 'Model-C'
```

```python
train_run_c = auto_trainer.run(
        inputs={"dataset": dataset_c.uri},
        params = {
            "model_class": 'sklearn.ensemble.RandomForestClassifier',
            "label_columns": "label",
            "model_name": model_name,
            "labels" : {'Status':'Challenger'},
            "tag" : "v1"
        }, 
        handler='train',
        local=False
    )
```

### Batch inference for the models

```python
prediction_set=pd.read_csv('./data/predict_data.csv')
prediction_set=project.log_dataset('prediction_set',df=prediction_set)
```

```python
print("Prediction Set")
prediction_set.to_dataitem().as_df()
```

```python
batch_inference_function = mlrun.code_to_function("batch_inference",kind='job',image='mlrun/mlrun',filename='./code/batch_inference.py')
```

```python
batch_inference_run = batch_inference_function.run(
    handler="infer",
    inputs={"dataset": prediction_set.uri},
    params={
        "model": train_run_a.outputs['model'],
        "label_columns" : "label",
        "result_set_name" : f"prediction_a1" 
    },
    local=False,
)
```

Saving Batch ID for others models predicitons - need to use the same batch_id number to compare between the models

```python
Batch_ID=batch_inference_run.metadata.labels['batch_id']
print(f"Batch ID --> {Batch_ID}")
```

```python
batch_inference_run = batch_inference_function.run(
    handler="infer",
    inputs={"dataset": prediction_set.uri},
    params={
        "model": train_run_b.outputs['model'],
        "label_columns": "label",
        "batch_id":Batch_ID,
        "result_set_name" : f"prediction_b1" 
        
    },
    local=False,
)
```

```python
batch_inference_run = batch_inference_function.run(
    handler="infer",
    inputs={"dataset": prediction_set.uri},
    params={
        "model": train_run_c.outputs['model'],
        "label_columns": "label",
        "batch_id":Batch_ID,
        "result_set_name" : f"prediction_model_c_1" 
        
    },
    local=False,
)
```

```python
y_true=pd.read_csv('./data/df_true_y.csv')
y_true=project.log_dataset('y_true',df=y_true)
```

```python
compare_models = mlrun.code_to_function("compare_models",kind='job',image='mlrun/mlrun',filename='./code/test_compare.py')
```

```python
compare_models_run=compare_models.run(handler='compare_models',
                  inputs={'y_true':y_true.uri},
                  params={'metrics':['accuracy','f1_score'],
                         'batch_id':Batch_ID,
                         'switch_model' :False,
                         'label_columns':'label',
                         'champion_metric' : 'accuracy',
                         'weights':{'accuracy' : 0.6}},local=True)
```

Saving Batch ID for others models predicitons - need to use the same batch_id number to compare between the models

```python
compare_id=compare_models_run.metadata.labels['comparsion_id']
print(f"Compare ID --> {compare_id}")
```
